```python
import tensorflow as tf
import base64
import cv2
import numpy as np
from flask import Flask,request
from tensorflow.keras.preprocessing import image
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
app = Flask(__name__)
new_model = tf.keras.models.load_model("C:/Users/Mahmaoud/first/conv2d-colonancer-99.99.h5",compile=False)
new_model.load_weights('C:/Users/Mahmaoud/first/conv2d-colonancer-weights.h5')
new_optimizer = tf.keras.optimizers.Adam(learning_rate= 0.001)
new_model.compile(optimizer=new_optimizer, loss='categorical_crossentropy')

def create_model_data1(image_path, batch_size):
    '''
    This function takes a single image and fits it into an image data generator, as the model takes data from the image data generator.
    The image data generator converts the image into a tensor.
    '''

    # Define model parameters
    img_size = (224, 224)
    channels = 3  # Either BGR or Grayscale
    color = 'rgb'
    img_shape = (img_size[0], img_size[1], channels)

    # This function will be used in the image data generator for data augmentation. It just takes the image and returns it again.
    def scalar(img):
        return img

    # Create an image data generator
    data_gen = ImageDataGenerator(preprocessing_function=scalar)

    # Load the image
    img = image.load_img(image_path, target_size=img_size)

    # Convert the image to a tensor
    img_tensor = image.img_to_array(img)
    img_tensor = np.expand_dims(img_tensor, axis=0)

    # Generate batches of the image tensor
    image_gen = data_gen.flow(img_tensor, batch_size=batch_size)

    return image_gen
@app.route('/api',methods = ['Put'] )
def index():
       inputchar = request.get_data()
       imgdata = base64.b64decode(inputchar)
       filename = 'somthing.jpg'  
       with open(filename, 'wb') as f:
        f.write(imgdata)
       image_path = "C:/Users/Mahmaoud/first/somthing.jpg" 
       im=create_model_data1(image_path,1)
       preds = new_model.predict(im)
       print(preds)
       y_pred = np.argmax(preds, axis=1)
       print(y_pred)
       result = ""
       if y_pred== 4: 
              result= 'lung_scc (Squamous Cell Carcinoma)'
       elif y_pred==0:
               result= 'Colon ACA (malignant)'
       elif y_pred==1:
               result= 'Colon_Benign cancer'
       elif y_pred==2:
              result='lung_aca (malignant)'
       elif y_pred==3:
              result='Lung_Benign cancer'     
       else:
              result='no'
              

    


       return result
       
        
if __name__ == "__main__":
    app.run(debug=True)
```
